```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# Configure matplotlib for publication-quality figures
mpl.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 12,
    "axes.titlesize": 14,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "figure.dpi": 300,
    "figure.figsize": (6, 3),
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.alpha": 0.6
})

# Data
angles = ["90–22", "45–22", "0–68", "135–45", "90–45", "0–112",
          "90–112", "45–68", "135–22", "90–68", "135–112", "45–112"]
errors = [13, 1, 8, 6, 1, 5, 3, 4, 2, 1, 1, 3]

# Create the plot
fig, ax = plt.subplots()
bars = ax.bar(angles, errors, color="steelblue", edgecolor="black")

# Labeling
ax.set_xlabel("Measurement Angle Pairs (Alice–Bob)")
ax.set_ylabel("Error Count")
ax.set_title("Error Distribution by Measurement Angles")

# Improve readability
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Save as high-resolution image suitable for Springer submission
plt.savefig("error_distribution_publication.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
blocks = [10,20,30,40,50,60,70,80,90]
qber = [0.8,0.7,0.6,0.1,0.5,0.3,0.5,0.5,0.5]

plt.plot(blocks, qber, marker='o')
plt.xlabel("Measurement Number")
plt.ylabel("QBER (%)")
plt.title("QBER Evolution Across Measurements")
plt.grid(True)
plt.savefig("qber_evolution.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    'font.family': 'DejaVu Sans',  # <- changed from 'Arial'
    'font.size': 11,
    'axes.labelsize': 11,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.figsize': (3.3, 2.5),
    'figure.dpi': 300,
})

labels = ['Classical\nBound', 'Measured S', 'Quantum\nBound']
values = [2.0, 0.022, 2.828]
colors = ['lightblue', 'red', 'lightgreen']

fig, ax = plt.subplots()
bars = ax.bar(labels, values, color=colors, edgecolor='black', alpha=0.8)

for bar, value in zip(bars, values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.1,
            f'{value}', ha='center', va='bottom', fontsize=9)

ax.axhline(y=2, color='gray', linestyle='--', alpha=0.7, linewidth=0.8)
ax.axhline(y=2.828, color='gray', linestyle='--', alpha=0.7, linewidth=0.8)

ax.set_ylabel("CHSH Value (S)")
ax.set_title("CHSH Inequality Test")
ax.set_ylim(0, 3.2)

plt.tight_layout()
plt.savefig("chsh_test_springer.png", dpi=300, bbox_inches="tight")
plt.savefig("chsh_test_springer.pdf", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update({
    'font.family': 'DejaVu Sans',  # <- safe font
    'font.size': 11,
    'axes.titlesize': 13,
    'figure.figsize': (3.3, 2.5),
    'figure.dpi': 300,
})

sizes = [94, 6]
labels = ['Discarded', 'Sifted']
colors = ['lightcoral', 'lightgreen']
percentages = [f'{s}%' for s in sizes]
legend_labels = [f'{label} ({pct})' for label, pct in zip(labels, percentages)]

fig, ax = plt.subplots()

wedges, texts = ax.pie(
    sizes,
    colors=colors,
    startangle=90,
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}
)

# Use legend instead of labels on pie
ax.legend(
    wedges, legend_labels, title="Category",
    loc="center left", bbox_to_anchor=(0.9, 0, 0.5, 1)
)

ax.set_title("Sifting Efficiency")

plt.tight_layout()
plt.savefig("sifting_efficiency_legend.png", dpi=300, bbox_inches="tight")
plt.savefig("sifting_efficiency_legend.pdf", bbox_inches="tight")
plt.show()
```
